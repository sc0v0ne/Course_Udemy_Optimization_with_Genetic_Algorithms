```python
import numpy as np
from geneticalgorithm import geneticalgorithm as ga
```

```python
# Define the equations

def fitness_function(x):
    x1 = x[0]
    x2 = x[1]
    x3 = x[2]


    # Apply Constraints

    penalty = 0

    if 5*x1 + 7*x2 + 4*x3 > 10:
        penalty = np.inf

    # Negative the objective function for maximization and apply penalty
    return -(16*x1+22*x2+12*x3) + penalty
```

```python
model = ga(
    function=fitness_function,
    dimension=3,
    variable_type='bool'
)

model.run()
```
